# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados nesta análise foram obtidos no endereço https://github.com/owid/covid-19-data/tree/master/public/data, em 09/09/2024. A principal fonte do *dataset* é a Organização Mundial da Saúde (OMS) e a plataforma de publicações Our World in Data (OWID), mas também são utilizadas informações provenientes de trabalhos acadêmicos, dos relatórios divulgados pelos governos locais, do Centro de Controle e Prevenção de Doenças e dos indicadores de desenvolvimento do Banco Mundial.

Os dados são apresentados por data e localidade e trazem informações sobre número de casos e óbitos, hospitalização e imunização. Apesar de serem extraídos de fontes oficiais, é importante lembrar que os dados podem não refletir de forma adequada a situação de determinado país. Isso se deve, entre outros fatores, à subnotificação, aos atrasos no processo de comunicação e às inconsistencias entre os dados nacionais consolidados e os obtidos de entes subnacionais.

A última hipótese tem grande impacto no Brasil, por sua vasta dimensão territorial e diferenças regionais nas condições de estrutura hospitalar e de comunicação. Além disso, nos meses iniciais da pandemia de COVID-19, as mudanças de orientação quase que diárias podem ter influenciado na forma como as informações sobre a doença eram repassadas pelas autoridades e unidades locais de saúde.

### Dicionário de variáveis

A seguir são apresentadas as variáveis correspondentes às colunas do *dataset*, agrupadas por tema. O *codebook* original pode ser obtido no endereço https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-codebook.csv (acesso em 09/09/2024). **Com exceção das variáveis pertencentes ao grupo "Testes", que deixaram de ser atualizadas em 23/06/2022, e ao grupo "Outras variáveis", que representam parâmetros fixos no período de um ano, as demais são atualizadas diariamente**.

**Casos confirmados**
*   `total_cases`: Total de casos confirmados de COVID-19. Pode incluir casos prováveis, quando reportados.
* `new_cases`: Novos casos confirmados de COVID-19. Pode incluir casos prováveis, quando reportados. Nos casos raros em que fonte relata uma mudança diária negativa devido a uma correção de dados, é apresentado o valor "NA".
* `new_cases_smoothed`: Novos casos confirmados de COVID-19 (média móvel de 7 dias). Pode incluir casos prováveis, quando reportados.
* `total_cases_per_million`: Total de casos confirmados de COVID-19 a cada 1.000.000 de pessoas. Pode incluir casos prováveis, quando reportados.
* `new_cases_per_million`: Novos casos confirmados de COVID-19 a cada 1.000.000 de pessoas. Pode incluir casos prováveis, quando reportados.
* `new_cases_smoothed_per_million`: Novos casos confirmados de COVID-19 (média móvel de 7 dias) a cada 1.000.000 de pessoas. Pode incluir casos prováveis, quando reportados

**Mortes confirmadas**
* `total_deaths`: Total de mortes atribuídas à COVID-19. Pode incluir mortes prováveis, quando reportadas.
* `new_deaths`: Novas mortes atribuídas à COVID-19. Pode incluir mortes prováveis, quando reportadas. Nos casos raros em que fonte relata uma mudança diária negativa devido a uma correção de dados, é apresentado o valor "NA".
* `new_deaths_smoothed`: Novas mortes atribuídas à COVID-19 (média móvel de 7 dias). Pode incluir mortes prováveis, quando reportadas.
* `total_deaths_per_million`: Total de mortes atribuídas à COVID-19 a cada 1.000.000 de pessoas. Pode incluir mortes prováveis, quando reportadas.
* `new_deaths_per_million`: Novas mortes atribuídas à COVID-19 a cada 1.000.000 de pessoas. Pode incluir mortes prováveis, quando reportadas.
* `new_deaths_smoothed_per_million`: Novas mortes atribuídas à COVID-19 (média móvel de 7 dias) a cada 1.000.000 de pessoas. Pode incluir mortes prováveis, quando reportadas.

**Excesso de mortalidade**
* `excess_mortality`: Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020-2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative`: Diferença percentual entre o número acumulado de mortes desde 01/01/2020 e o número acumulado projetado de mortes para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative_absolute`: Diferença acumulada entre o número relatado de mortes desde 01/01/2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative_per_million`: Diferença acumulada entre o número relatado de mortes desde 01/01/2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.

**Hospitais e UTI**
* `icu_patients`: Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia.
* `icu_patients_per_million`: Número de pacientes com COVID-19 em UTIs em um determinado dia por 1.000.000 de pessoas.
* `hosp_patients`: Número de pacientes com COVID-19 no hospital em um determinado dia.
* `hosp_patients_per_million`:	Número de pacientes com COVID-19 no hospital em um determinado dia a cada 1.000.000 de pessoas.
* `weekly_icu_admissions`:	Número de pacientes com COVID-19 recém-admitidos em UTIs em uma semana (data de relatório e 6 dias anteriores).
* `weekly_icu_admissions_per_million`:	Número de pacientes com COVID-19 recém-admitidos em UTIs em uma semana a cada 1.000.000 de pessoas (data de relatório e 6 dias anteriores).
* `weekly_hosp_admissions`:	Número de pacientes com COVID-19 recém-admitidos em hospitais em uma semana (data de relatório e 6 dias anteriores).
* `weekly_hosp_admissions_per_million`:	Número de pacientes com COVID-19 recém-admitidos em hospitais em uma semana a cada 1.000.000 de pessoas (data de relatório e 6 dias anteriores).

**Respostas políticas**
* `stringency_index`: Trata-se do Índice de Rigor da Resposta Governamental, que é uma medida baseada em nove indicadores de resposta, como fechamento de escolas e de locais de trabalho e restrições de viagem. Reescalada para um valor de 0 a 100, em que 100 é a resposta mais rigorosa.

**Taxa de reprodução efetiva**
* `reproduction_rate`:	Estimativa em tempo real da taxa de reprodução efetiva da COVID-19, que reflete o número de casos de contaminação secudária em função de uma contaminação primária.

**Testes**
* `total_tests`:	Total de testes para COVID-19.
* `new_tests`:	Novos testes para COVID-19 (calculado apenas para dias consecutivos).
* `total_tests_per_thousand`: Total de testes para COVID-19 para cada 1.000 pessoas.
* `new_tests_per_thousand`: Novos testes para COVID-19 para cada 1.000 pessoas.
* `new_tests_smoothed`:	Novos testes para COVID-19 (média móvel de 7 dias). Para países que não relatam dados de testes diariamente, assume-se que a quantidade de testes mudou uniformemente nos dias sem dados relatados, o que leva a uma série completa de números diários, usada para computar a média móvel de 7 dias.
* `new_tests_smoothed_per_thousand`: Novos testes para COVID-19 (média móvel de 7 dias) a cada 1.000 pessoas.
* `positive_rate`: Percentual de testes de COVID-19 que são positivos, obtido a partir da média móvel de 7 dias (i.e., o inverso de `tests_per_case`).
* `tests_per_case`:	Testes realizados para cada novo caso confirmado de COVID-19, obtido a partir da média móvel de 7 dias (i.e., `positive_rate`).
* `tests_units`:	Unidades usadas pela localidade para relatar seus dados de teste. Um arquivo de um país não pode conter unidades diferentes, ou seja, todas as métricas relacionadas aos dados de teste, para um mesmo país, utilizam mesma a unidade de teste especificada. São unidades válidas para esse fim: "pessoas testadas" (número de pessoas testadas), "testes realizados" (número de testes realizados) e "amostras testadas" (número de amostras testadas). Essas unidades não são equivalentes, visto que uma mesma pessoa pode ser testada mais de uma vez em um dia e pode ser necessária mais de uma amostra para realizar um teste.

**Vacinações**
* `total_vaccinations`:	Total de doses de vacina contra COVID-19 administradas.
* `people_vaccinated`: Total de pessoas que receberam pelo menos uma dose da vacina.
* `people_fully_vaccinated`: Total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação.
* `total_boosters`:	Total de doses de reforço da vacina contra COVID-19 administradas (i.e., além do número prescrito pelo protocolo inicial de vacinação).
* `new_vaccinations`: Novas doses de vacina contra COVID-19 administradas (calculada apenas para dias consecutivos).
* `new_vaccinations_smoothed`:	Novas doses de vacina contra COVID-19 administradas (média móvel de 7 dias). Para países que não relatam dados de testes diariamente, assume-se que a quantidade de doses administradas mudou uniformemente nos dias sem dados relatados, o que leva a uma série completa de números diários, usada para computar a média móvel de 7 dias.
* `total_vaccinations_per_hundred`: Total de doses de vacina contra COVID-19 administradas a cada 100 pessoas da população total.
* `people_vaccinated_per_hundred`:	Total de pessoas que receberam pelo menos uma dose de vacina a cada 100 pessoas da população total.
* `people_fully_vaccinated_per_hundred`:	Total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação a cada 100 pessoas da população total.
* `total_boosters_per_hundred`:	Total de doses de reforço da vacina contra COVID-19 administradas a cada 100 pessoas da população total.
* `new_vaccinations_smoothed_per_million`:	Novas doses de vacina contra COVID-19 administradas (média móvel de 7 dias) a cada 1.000.000 de pessoas da população total.
* `new_people_vaccinated_smoothed`:	Número diário de pessoas recebendo sua primeira dose de vacina (média móvel de 7 dias).
* `new_people_vaccinated_smoothed_per_hundred`:	Número diário de pessoas recebendo sua primeira dose de vacina (média móvel de 7 dias) a cada 100 pessoas da população total.

**Outras variáveis**
* `iso_code`: ISO 3166-1 alpha-3 – códigos de três letras que identifica o país a que os dados se referem, de acordo com a norma ISO 3166-1 alpha 3. No caso de regiões definidas pelo Our World in Data (OWID), página que pública parâmetros socioeconômicos de vários países, essa variável inclui o prefixo 'OWID_' (por exemplo, `OWID_World` ou `OWID_Europe`).
* `continent`:	Continente da localização geográfica.
* `location`:	Localização geográfica.
* `date`:	Data de observação.
* `population`:	População total (dados mais recentes disponíveis).
* `population_density`:	Densidade populacional em pessoas por quilômetro quadrado (dados mais recentes disponíveis).
* `median_age`:	Mediana da idade da população (projeção feita pela ONU para o ano de 2020).
* `aged_65_older`:	Percentual da população com 65 anos ou mais (dados mais recentes disponíveis).
* `aged_70_older`:	Percentual da população com 70 anos ou mais (dados de 2015).
* `gdp_per_capita`:	Produto Interno Bruto (PIB) per capita em paridade do poder de compra em dólares internacionais  (dados mais recentes disponíveis).
* `extreme_poverty`:	Percentual da população vivendo em extrema pobreza (dados mais recentes disponíveis a partir de 2010).
* `cardiovasc_death_rate`:	Taxa de mortalidade por doenças cardiovasculares a cada 100.000 pessoas (dados de 2017).
* `diabetes_prevalence`:	Prevalência de diabetes, dado pelo percentual da população com idade entre 20 e 79 anos que possui a doença (dados de 2017).
* `female_smokers`:	Percentual de mulheres fumantes (dados mais recentes disponíveis).
* `male_smokers`:	Percentual de homens fumantes (dados mais recentes disponíveis).
* `handwashing_facilities`:	Percentual da população que conta com instalações básicas de lavagem das mãos nas imediações (dados mais recentes disponíveis).
* `hospital_beds_per_thousand`:	Leitos hospitalares para cada 1.000 pessoas (dados mais recentes disponíveis a partir de 2010).
* `life_expectancy`:	Expectativa de vida ao nascer em 2019.
* `human_development_index`:	Índice de Desenvolvimento Humano (IDH) (dados de 2019).

### Importando os dados do COVID-19

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

## Análise Exploratória dos Dados

A seguir serão observados aspectos gerais do *dataset*, como entradas iniciais, tamanho e percentual de campos nulos. Também serão verificados os países com mais casos e com mais mortes, a evolução do numero de casos ao longo do tempo e a relação entre o PIB per capita e número de mortes.



```
# visualizando as 5 primeiras entradas
df.head()
```

```
# tamanho do data-frame, tipos das variáveis e nomes das colunas
print("Tamanho do data-frame: ",df.shape,"\n\n")
print("Tipo das variáveis:\n",df.dtypes,"\n\n")
print("Nome das colunas:\n" , df.columns)
```

```
# % de valores ausentes em ordem decrescente
(df.isnull().sum() / df.shape[0]).sort_values(ascending=False).head(15)
```

```
# transformando a coluna date no formato datetime
df.date = pd.to_datetime(df.date)
```

```
# verificando os países com mais casos
df.loc[(df.date == '2022-06-23') & (df.iso_code.str[0:5] != "OWID_")].sort_values(by="total_cases", ascending=False)
```

```
# verificando os países com mais mortes
df.loc[(df.date == '2022-06-23') & (df.iso_code.str[0:5] != "OWID_")].sort_values(by="total_deaths", ascending=False)
```

```
# plotando gráfico de barras para os 5 países com mais mortes
top5_total_deaths = df.loc[(df.date == '2022-06-23') & (df.iso_code.str[0:5] != "OWID_"), ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]
top5_total_deaths.total_deaths = top5_total_deaths.total_deaths / 1e6
fig1, ax1 = plt.subplots(figsize=(5,5))
ax1.bar('location', 'total_deaths', data=top5_total_deaths, color='black')
ax1.set_title("Países com mais mortes (até 23/06/2022)")
ax1.set_ylabel("Número de mortes (milhões)");
```

```
# plotando gráfico de linhas para ver a evolução do total de casos de COVID-19 no mundo
dados_para_plotar = df.loc[df.location == "World", ["date", "total_cases"]]
dados_para_plotar.total_cases = dados_para_plotar.total_cases / 1e6
# mudando o índice para a data para que o gráfico mostre a evolução em função dos anos
dados_para_plotar.set_index("date", inplace=True)
fig2, ax2 = plt.subplots(figsize=(5,5))
dados_para_plotar.plot(ax=ax2,kind="line",color="black")
ax2.set_title("Evolução do número de casos")
ax2.set_xlabel("Ano");
ax2.set_ylabel("Número de casos (milhões)");
```

```
# plotando gráfico de PIB per capita por número de mortes a cada um milhão de habitantes
dados_para_plotar = df.loc[(df.date == '2022-06-23') & (df.iso_code.str[0:5] != "OWID_"),["gdp_per_capita", "total_deaths_per_million"]]
dados_para_plotar.total_deaths = dados_para_plotar.total_deaths_per_million
fig3, ax3 = plt.subplots(figsize=(5,5))
ax3.scatter(dados_para_plotar.gdp_per_capita, dados_para_plotar.total_deaths_per_million,s=4,color="black")
ax3.set_title("Relação entre PIB per capita e mortes")
ax3.set_xlabel("PIB per capita (dólares)")
ax3.set_ylabel("Número de mortes a cada um milhão de habitantes");
```

### Análise Exploratória para o Brasil

Feita a análise geral dos dados a nível mundial, serão apresentadas informações sobre a COVID-19 a nível nacional.

```
# identificando as datas de registro da primeira morte e do primeiro caso no Brasil e a diferença de dias entre elas
df_brasil = df.loc[df.location == "Brazil"].copy()
data_1a_morte = df_brasil.loc[df_brasil.total_deaths >= 1].sort_values(by="date").reset_index(drop=True).loc[0, "date"]
data_1o_registro = df_brasil[df_brasil.total_cases >= 1].sort_values(by="date").reset_index(drop=True).loc[0, "date"]
print("A primeira morte foi registrada no Brasil em", data_1a_morte.strftime("%d/%m/%Y"),".")
print("O primeiro caso de COVID-19 foi registrado no Brasil em", data_1o_registro.strftime("%d/%m/%Y"),".")
print("Passaram-se",(data_1a_morte-data_1o_registro).days, "dias entre o primeiro caso e a primeira morte.")
```

```
# plotando gráfico de linhas para ver a evolução do total de casos e de mortes no Brasil
dados_para_plotar_brasil = df_brasil[["date", "total_cases","total_deaths"]].copy()
dados_para_plotar_brasil.total_cases = dados_para_plotar_brasil.total_cases / 1e6
dados_para_plotar_brasil.total_deaths = dados_para_plotar_brasil.total_deaths / 1e6
# mudando o índice para a data para que o gráfico mostre a evolução em função dos anos
dados_para_plotar_brasil.set_index("date", inplace=True)
fig3, (ax3, ax4) = plt.subplots(nrows=1, ncols=2, figsize=(10,5),sharex=True)
dados_para_plotar_brasil.plot(ax=ax3,kind="line")
ax3.set_title("Evolução de casos e mortes no Brasil (linear)")
ax3.set_xlabel("Ano");
ax3.set_ylabel("Número de casos ou mortes (milhões)");
dados_para_plotar_brasil.plot(ax=ax4,kind="line")
ax4.set_yscale('log')
ax4.set_title("Evolução de casos e mortes no Brasil (logarítmica)")
ax4.set_xlabel("Ano");
ax4.set_ylabel("Número de casos ou mortes (milhões)");
```

## Conclusão

A análise exploratória inicial dos dados revelou que as variáveis relativas aos grupo "Hospitais e UTI" são as que mais apresentam campos não preenchidos, o que provavelmente se deve ao fato de esses valores dependerem de relatórios de governos locais, que, por sua vez, baseiam-se em informações dispersas das ínumeras unidades hospitalares em seu território. A dificuldade em mater atualizados e consolidados dados tão pulverizados pode ter contribuído para a incompletude as variáveis desse grupo. As variáveis do grupo "Excesso de mortalidade" também apresentam alto percentual de valores nulos, mas não foi identificada uma razão para isso.

Os países com mais mortes foram, em ordem decrescente, Estados Unidos, Brasil, Índia, Rússia e México. Ao analisar o número total de casos no mundo ao longo dos anos, verifica-se que essa grandeza aumentou mais de sete vezes entre 2021 e 2023, tornando-se estável após esse período.

O gráfico de dispersão que relaciona o número total de mortes a cada um milhão de habitantes com o PIB per capita indica que países com menor PIB apresentam maior incidência de casos fatais, o que é esperado tendo em vista que o PIB, ao refletir a riqueza de uma nação, também sinaliza, ainda que indiretamente, os recursos que o governo e a população têm disponíveis para ações de prevenção e tratamento de saúde. Nesse sentido, faz sentido que os habitantes de países mais pobres seja mais impactado pela doença. Vale lembrar, no entanto, que essa análise não levou em conta fatores como imunização e medidas de isolamento, o que poderia levar a conclusões mais detalhadas sobre o tema.

A análise dos dados no Brasil mostrou que o primeiro caso foi registrado em 01/03/2020, e a primeira morte 21 dias depois, em 22/03/2020. Os gráficos mostrando a evolução do número de casos e do número de mortes seguiu a tendencia daquele obtido na análise feita para os dados a nível mundial, com um rápido crescimento entre 2021 e 2023, seguida pela estabilização nos meses seguintes ao período.
